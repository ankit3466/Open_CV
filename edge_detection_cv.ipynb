```
import cv2
from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('tetris_blocks.png')
```

```
cv2_imshow(img)
```

```
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
```

```
cv2_imshow(gray)
```

```
gray.shape
```

```
# edge detection
edg = cv2.Canny(gray,75,200)
```

```
cv2_imshow(edg)
```

```
edg1 = cv2.Canny(img ,5,20)
cv2_imshow(edg1)
```

```
# Draw counters
cnt = cv2.findContours(edg , cv2.cv2.RETR_LIST , cv2.CHAIN_APPROX_SIMPLE)
```
