```
import cv2
```

```
img = cv2.imread('dog.jpg')     # reading image
```

```
# cv2.imshow() -----> not supported in colob
from google.colab.patches import cv2_imshow
```

```
# showing image
cv2_imshow(img)
```

```
# shape of original image
img.shape
```

```
# cv2 function to resize a image
# It change the pixcles
x = cv2.resize(img,(100,150))
```

```
# shape of resized image
x.shape
```

```
cv2_imshow(x)
```

```
# another example of resizing
y=cv2.resize(img,(8,8))
```

```
cv2_imshow(y)
```

```
import matplotlib.pyplot as plt
plt.imshow(y)
```

```
# Rotation of image


# taking row and columns

rows,cols = img.shape[:2]
```

```
# checking the centre of image 
# from centre it rotate the image

M = cv2.getRotationMatrix2D(((cols-1)/2.0,(rows-1)/2.0),90,1)
```

```
# wrapping up because rows and colmns are diff
dst = cv2.warpAffine(img,M,(cols,rows))
```

```
cv2_imshow(dst)
```
